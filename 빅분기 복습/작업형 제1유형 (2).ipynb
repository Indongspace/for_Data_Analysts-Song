```python
# 10. 주어진 y컬럼의 모든 데이터를 활용하여 1사분위수와 3사분위수 차이를 절대값으로 구하고 구한 절대값의 소수점을 버리고 정수로 출력하시오.
# (단, 출력화면에는 정답인 정수만 출력)
import numpy as np
import pandas as pd
x = np.random.choice(range(1,21), size=10, replace=False)
y = [1,2,3,4,5,37,41,42,44,10]
a = pd.DataFrame({'y':y, 'x':x})
quantiles = a['y'].quantile([0.25, 0.5, 0.75])
quantiles = pd.DataFrame({'quantiles':quantiles})
Q1 = quantiles.iloc[0,0]
Q3 = quantiles.iloc[2,0]
int(np.floor(abs(Q1 - Q3)))
```
