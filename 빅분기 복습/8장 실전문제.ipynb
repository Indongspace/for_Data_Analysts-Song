```python
# 01. 서울특별시 대기환경정보 데이터를 이용해서 서울시의 미세먼지, 초미세먼지를 분석한다.
import pandas as pd
airseoul = pd.read_excel(r"D:\2025 단축키 빅데이터분석기사_파이썬\작업형기초\data\period1.xlsx")
airseoul
```

```python
# 04. 서울의 음식점에 관한 데이터이다.
import pandas as pd
food = pd.read_csv(r"D:\2025 단축키 빅데이터분석기사_파이썬\작업형기초\data\6110000_서울특별시_07_24_04_P_일반음식점.csv", encoding='cp949',
                  na_values='', low_memory=False)
```

```python
food1 = food.rename(columns={
    '인허가일자':'open_date',
    '상세영업상태명':'status',
    '폐업일자':'close_date',
    '사업장명':'name',
    '업태구분명':'type',
    '소재지전체주소':'address'
})
food1 = food1[['open_date','status','name','close_date','type','address']]
food1
```

```python
food1.info()
```

```python
food1.isna().sum()
```

```python
food1['type'].value_counts()
```

```python
food1['type'] = food1['type'].replace(['까페','다방','라이브카페','커피숍','카페'], '카페')
```

```python
food1['type'] = food1['type'].replace(['통닭(치킨)','호프/통닭'], '치킨')
```

```python
food1['type'] = food1['type'].replace(['일식','회집','횟집'],'회집')
```

```python
food1['type'] = food1['type'].replace(['경양식','패밀리레스토랑'], '레스토랑')
food1['type'] = food1['type'].replace(['정종/대포집/소주방'], '소주방')
food1['type'] = food1['type'].replace(['외국음식전문점(인도,태국등)'], '외국음식전문점')
food1['type'] = food1['type'].replace(['기타','193959.1505'], pd.NA)
```

```python
food1['open_date'].min(), food1['open_date'].max()
```

```python
food1.loc[food1['open_date'] < food1['open_date'].min(), 'open_date'] = None
```

```python
food1
```

```python
food1['open_date'] = food1['open_date'].astype(str)
food1['open_year'] = food1['open_date'].str.slice(0,4)
```

```python
food1
```

```python
food1['close_date'] = food1['close_date'].astype(str)
food1['close_year'] = food1['close_date'].str.slice(0,4)
food1
```

```python
# (1) 가장 오래 영업 중인 음식점업의 open_year과 type, name을 출력하시오.
import pandas as pd
filtered_food1 = food1[(~food1['open_date'].isna()) & (food1['status']=='영업')]
filtered_food1
```

```python
filtered_food1 = filtered_food1[filtered_food1['open_date']==filtered_food1['open_date'].min()]
filtered_food1
```

```python
selected_columns = ['type','open_year','name']
filtered_food1 = filtered_food1[selected_columns]
print(filtered_food1)
```

```python
# (2) 개업한 전체 음식점에서 상위 3개 업종의 비율을 출력하시오.
filtered_food1 = food1[(~food1['open_date'].isna()) & (~food1['type'].isna())]
```

```python
filtered_food1
```

```python
type_counts = filtered_food1.groupby('type').size().reset_index(name='n')
type_counts
```

```python
type_counts['total'] = type_counts['n'].sum()
type_counts['pct'] = round(type_counts['n'] / type_counts['total'] * 100, 2)
type_counts
```

```python
type_counts = type_counts.sort_values(by='n', ascending=False)
top_3 = type_counts.head(3)
```

```python
print(top_3)
```
