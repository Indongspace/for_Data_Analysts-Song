```python
# 01. 서울특별시 대기환경정보 데이터를 이용해서 서울시의 미세먼지, 초미세먼지를 분석한다.
import pandas as pd
airseoul = pd.read_excel(r"D:\2025 단축키 빅데이터분석기사_파이썬\작업형기초\data\period1.xlsx")
airseoul
```

```python
airseoul.info()
```

```python
airseoul.columns
```

```python
airseoul1 = airseoul.rename(columns={
    '날짜':'date',
    '측정소명':'region',
    '미세먼지 PM10\n(㎍/m3)':'pm10',
    '초미세먼지\nPM2.5 (㎍/m3)':'pm2.5',
})
airseoul1 = airseoul1[['date','region','pm10','pm2.5']]
airseoul1
```

```python
airseoul1['region'].value_counts()
```

```python
airseoul1 = airseoul1[~((airseoul1['date']=='전체') | (airseoul1['region']=='평균'))]
airseoul1
```

```python
airseoul1.isna().sum()
```

```python
airseoul2 = airseoul1.dropna()
```

```python
# (1) 미세먼지가 가장 심했던 날짜, 지역, 미세먼지 수치(pm10)은?
airseoul2[airseoul2['pm10'] == airseoul2['pm10'].max()]
```

```python
# (2) 서울시의 2022년 2월부터 3월까지 미세먼지 평균을 구하고 평균기준으로 상위 5개 구는?
import pandas as pd
grouped_mean = airseoul2.groupby('region')['pm10'].mean().reset_index()
grouped_mean
```

```python
sorted_mean = grouped_mean.sort_values(by='pm10', ascending=False)
sorted_mean
```

```python
top_5 = sorted_mean.head(5)
```

```python
print(top_5)
```

```python
# (3) 하루의 미세먼지(pm10) 상황은 1일 수치가 0~30이면 'good', 31~80이면 'normal', 81~150이면 'bad', 151이상이면 'worse'구분한다. 
# 이 기준에 따라 미세먼지 등급 변수(파생변수)를 생성하고 이때 'good'의 빈도수와 백분율을 구하시오.
airseoul2
```

```python
airseoul2['pm_grade'] = pd.cut(airseoul2['pm10'], bins=[-float('inf'), 31, 81, 151, float('inf')], labels=['good','normal','bad','worse'], right=False)
airseoul2
```

```python
airseoul2.loc[:,'pm_grade'] = pd.cut(airseoul2['pm10'], bins=[-float('inf'), 31, 81, 151, float('inf')], labels=['good','normal', 'bad', 'worse'], right=False)
airseoul2
```

```python
airseoul2.loc[:,'pm_grade'] = pd.cut(airseoul2['pm10'], bins=[-float('inf'), 31, 81, 151, float('inf')], labels=['good', 'normal', 'bad', 'worse'], right=False)
airseoul2
```

```python
grouped = airseoul2.groupby('pm_grade').size().reset_index(name='n')
grouped
```

```python
grouped['total'] = grouped['n'].sum()
grouped
```

```python
grouped['pct'] = grouped['n'] / grouped['total'] * 100
print(grouped)
```

```python
grouped[grouped['pm_grade']=='good'][['total','pct']]
```

```python
grouped.loc[grouped['pm_grade']=='good', ['total','pct']]
```

```python
print(grouped.loc[grouped['pm_grade']=='good', 'total'])
```
