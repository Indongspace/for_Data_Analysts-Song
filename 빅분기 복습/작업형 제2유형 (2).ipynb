```python
import pandas as pd
bike_df = pd.read_csv(r"D:\2025 단축키 빅데이터분석기사_파이썬\작업형 제2유형\data\bike_train.csv")
print(bike_df.head())
```

```python
import datetime as dt
bike_df['datetime'] = pd.to_datetime(bike_df['datetime'])
```

```python
bike_df['year'] = bike_df['datetime'].dt.year
bike_df['month'] = bike_df['datetime'].dt.month
bike_df['day'] = bike_df['datetime'].dt.day
bike_df['hour'] = bike_df['datetime'].dt.hour
bike_df.head()
```

```python
drop_columns = ['datetime', 'casual', 'registered']
bike_df = bike_df.drop(columns=drop_columns)
bike_df.head(3)
```

```python
bike_df.isna().sum()
```

```python
Y = bike_df['count']
Y.shape
```

```python
X = bike_df.drop(['count'], axis=1)
X.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
y_pred = reg.predict(x_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred, squared=False)
```

```python
x_train = pd.get_dummies(x_train, columns=['brand','transmission','fuelType'])
x_train = x_train.drop(['model'], axis=1)
print(x_train.head(3))
```

```python
x_test = pd.get_dummies(x_test, columns=['brand','transmission','fuelType'])
x_test = x_test.drop(columns=['model'])
```

```python
x_test.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train_carid = x_train['carID']
x_train_drop = x_train.drop(columns=['carID'])
x_train_normal = pd.DataFrame(scaler.fit_transform(x_train_drop), columns=x_train_drop.columns)
x_train_normal['carID'] = x_train_carid

x_test_carid = x_test['carID']
x_test_drop = x_test.drop(columns=['carID'])
x_test_normal = pd.DataFrame(scaler.transform(x_test_drop), columns=x_test_drop.columns)
x_test_normal['carID'] = x_test_carid
```

```python
x_train_normal.head(3)
```

```python
x_test_normal.head(3)
```

```python
train = pd.merge(x_train_normal, y_train, on='carID', how='left')
```

```python
train.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
train_data, valid_data = train_test_split(train, test_size=0.3, random_state=42)
```

```python
valid_data.head()
```

```python
pred_rf1 = model_rf1.predict(valid_data.drop(columns=['price','carID']))
```
